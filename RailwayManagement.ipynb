```python
import numpy as np
import pandas as pd

class Ticket_col:
    def __init__(self, name, Id, train_no):
        self.name = name
        self.Id = Id
        self.train_no = train_no

    def display_details(self):
        print("Name: ", self.name)
        print("TT ID: ", self.Id)
        print("Train No: ", self.train_no)


try:
    with open("/content/sample_data/HELLO!!!.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("The file does not exist.")

while True:
    try:
        n = int(input("Press 1 for entering TC Details\nPress 2 for Printing TC info\nPress 3 for Entering Train Data\nPress 4 for Printing Train data: "))

        if n == 1:
            m = int(input("Enter the number of TCs you want to enter: "))
            Info = []
            for i in range(m):
                name = input("Enter name:")
                Id = input("Id : ")
                train_no = int(input("Train No: "))
                abc = Ticket_col(name, Id, train_no)
                Info.append(abc)
            a = input("\nDo you want to perform more tasks: Y/N----> ").upper()
            if a== 'Y':
              continue
            elif a== 'N':
              break
            else:
              print("INVALID INPUT")

        if n == 2:
            if not Info:
                print("No TC details entered yet.")
            else:
                print("\nDetails of all TCs:")
                for abc in Info:
                    abc.display_details()
                a = input("\nDo you want to perform more tasks: Y/N----> ").upper()
                if a== 'Y':
                  continue
                elif a== 'N':
                  break
                else:
                  print("INVALID INPUT")

        if n == 3:
            df = pd.DataFrame(data=np.arange(0, 20).reshape(5, 4), index=["1.", "2.", "3.", "4.", "5."], columns=["Train No.", "    Train Name", "   Distance Covered", "   No_Of_Passengers"])
            for i in range(5):
                print(f"Enter Information for train {i}")
                for j in range(1):
                    df.iat[i, j] = input("Enter Train No: ")
                    j += 1
                    df.iat[i, j] = input("Enter Train Name: ")
                    j += 1
                    while True:
                        try:
                            df.iat[i, j] = int(input("Enter Distance Covered: "))
                            break
                        except ValueError:
                            print("Please enter a valid integer for Distance Covered.")
                    j += 1
                    while True:
                        try:
                            df.iat[i, j] = int(input("Enter No Of Passengers: "))
                            break
                        except ValueError:
                            print("Please enter a valid integer for No Of Passengers.")
            a = input("\nDo you want to perform more tasks: Y/N----> ").upper()
            if a== 'Y':
              continue
            elif a== 'N':
              break
            else:
              print("INVALID INPUT")

        if n == 4:
            if 'df' not in locals():
                print("No train data entered yet.")
            else:
                print(df)
                a = input("\nDo you want to perform more tasks: Y/N----> ").upper()
                if a== 'Y':
                  continue
                elif a== 'N':
                  break
                else:
                  print("INVALID INPUT")

    except ValueError:
        print("Please enter a valid integer for your choice.")
print("Task Completed Succesfully!")
```
